1)Demonstrate methods to handle missing data with the studentsâ€™performance datasets.

```python
import pandas as pd
import numpy as np
```

```python
students = pd.read_csv("students.csv")
students
```

```python
students.fillna(0)          #it will replace missing(NaN) values with 0
```

```python
students.fillna({'marks' : 'none'})   
#using this method, with the help of dictionary we can replace missing values with different values on the basis of columns
```

```python
students.fillna(method = 'ffill')            #forward fill 
#by default, axis=0(i.e. replaces value by previous row of same column)
#we can explicitly specify axis=1 if we want to replace missing value with the value of same row and previous column
```

```python
students.fillna(method = 'pad')              #it works same as ffill method
```

```python
students.fillna(method = 'bfill')         #backward fill
#by default, axis=0(i.e. replaces value by next row of same column)
#we can explicitly specify axis=1 if we want to replace missing value with the value of same row and next column
```

```python
students.fillna(0, limit=10)         
#by using this attribute we can limit number of replaces of missing value. It will fill only ten empty place by 0
```

```python
students.fillna(method='ffill', limit=10)         
#it will fill only empty ten place by its previous value 
```

```python
students.fillna(students['marks'].mean())       
#to fill missing values by specific column mean 
```

```python
students.fillna(students['marks'].median())
#to fill missing values by specific column mean 
```

4) Distinguish Pie chart,Bar  graph and  Histogram with  respect  to  their applicability in  data visualization.

Pie chart
- A pie chart shows a static number and how categories represent part of a whole-the composition of something.
- The entire pie would be 100 percent. 
- Pie charts focus on showing parts of a whole.
- A pie Graph can be either categorical or quantitative, but it is expressed as percent of the whole.

Bar chart
- Bar charts make comparing values easy.
- Often use a bar graph and column chart in the same way, though column charts limit your label and comparison space.
- The wide bars and segregated measurements emphasize the differences between values, rather than the flow of one value to another as a line graph would do.
- A bar graph can be used for either quantitative or categorical data. There should be small gaps between the bars.

Histogram
- Histograms categorize data by breaking it into bins, where each bin contains a subset of the data range.
- A histogram displays the number of items in each bin so that you can see the distribution of data and the progression of data from bin to bin.
- Histograms can be viewed in different ways by using its different types like bar, barstacked, stepfilled, step.
-  A histogram can only be used for quantitative data, there should not be gaps between bins, unless a bin has no data in it. 

5) Apply data visualization tool to highlight top 5 states of India according to population of the states. Consider  2011  year  to  find  top  5  states. Refer the  datasets Population_of_India-2001-2011.xls

```python
population = pd.read_csv('Population_of_India-2001-2011.csv')
population
```

```python
indexNames = population[ population['2011'] ==  1201810].index
population.drop(indexNames, inplace=True)
```

```python
population
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

plt.figure(figsize=(8,4))
population.sort_values('2011',axis=0,inplace=True, ascending=False)
names=population['State/Union Territory'].head()
values= population['2011'].head()
plt.bar(names, values)
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,4))
sns.barplot(x=names, y=values)
plt.title('Top 5 highly populated cities of India', fontsize=20)
plt.ylabel('Population', fontsize=15)
plt.xlabel('State/Union Territory',fontsize=15)
```
